```python
#Exercise 31
#Thomas Algorith
import matplotlib.pyplot as plt
import math

def thomas(a, b, c, d):
    """
      ALGORITHM DUE TO THOMAS:
    ARGUMENT:
        (array)a: array containing lower diagonal ( not used a[0])
        (array)b:array contains main diagonal
        (array)c: array contains upper diagonal (not used c[-1])
        (array)d: right hand side of the system
    RETURN:
        (array)x: Solution array of the system
    """

    N = len(b)
    X = [0]*N

    # ELIMINATION:
    for L in range(1,N):
        q = a[L]/b[L-1]
        b[L] = b[L] - c[L-1]*q
        d[L] = d[L] - d[L-1]*q
    # BACKWARD SUBSTITUTION:
    q = d[N-1]/b[N-1]
    X[N-1] = q

    for k in range(N-2,-1,-1):
        q = (d[k]-c[k]*q)/b[k]
        X[k] = q

    return X

#h =delta=X[i+1]-X[i]
k, h, u0, upiBI6, s = 9, 0.1, 11/9, 11/9, (22/7)
# where the interval is [0,m]
N = int(s/h) - 1 # the step count
# print(N)
a = [k*k*h*h - 2]*N
b = [1]*N
c = [1]*N
d = [2*k*k*h*h - u0] + [2*k*k*h*h]*(N-2) + [2*k*k*h*h - upiBI6]
u_sol = [u0] + thomas(a,b,c,d) + [upiBI6]
axis_x, u_exact = [0]*(N+2), [11/9]*(N+2)
for j in range(1, N+2) :
  axis_x[j] = axis_x[j-1] + h
  point_x = axis_x[j]
  u_exact[j] = math.cos(3*point_x) + math.sin(3*point_x) + 2/9

# WE WILL PRINT(u_sol)
# We will print(U_exact)
# We will print (axis_x)


figure = plt.subplot()
figure.set_title(' with u(0)=11/9,u(pi/6)=11/9,u" = k*k u + 2,')
figure.set_xlabel('x-axis')
figure.set_ylabel('u-axis')
figure.plot(axis_x, u_sol, 'go--', label='Solution Using Thomas Algorithm ')
figure.plot(axis_x, u_exact, 'ro--', label='(ExactSolution)')
figure.legend()
print('WE WILL ONLY CONSIDER FOR X-RANGE (0,pi/6)')




```
