```python
#Exercise 21
#Vector ClasS
class Vector :
  def __init__(self, n):
    self.vector = [0]*n

  def get_dimension(self) :
    return len(self.vector)

  def getVector(self):
    return self.vector

  def setVector(self, newVector):
    self.vector = newVector

  def add(self, newVector) :
    size = self.get_dimension()
    tmp = [0]*size
    for i in range(size) :
      tmp[i] = self.vector[i] + newVector[i]
    self.setVector(tmp)

  def scalarMultiply(self, scalar) :
    size = self.get_dimension()
    tmp = [0]*size
    for i in range(size) :
      tmp[i] = self.vector[i]*scalar
    self.setVector(tmp)



```

```python
#Testing
vecA = Vector(4)
vecB=Vector(4)
vecA.setVector([1,3,5,10])
vecB.setVector([2,4,6,10])

print('Printing The vector A: ')
print(vecA.getVector())
print('Printing The vector B: ')
print(vecB.getVector())

vecA.add([1,3,5,10])
print('Adding vector A with Itself we get: ')
print(vecA.getVector())
vecB.scalarMultiply(3)
print('Multiplying by a scalr 3 with B we get: ')
print(vecB.getVector())
```
