```python
#Exercise 27
#Forward Euler Method
import numpy as npy
import matplotlib.pyplot as plt
import math
f=lambda t:math.exp(-t)
inip=0 #initial point of the interval [0,2]
finp=2 #Final point of the interval[0,2]
delta_t=0.1
k=1
N_t=int((finp-inip)/delta_t)
t=npy.arange(0,2+delta_t,delta_t)
a=1
y=[]
yexact=[]
t0=0
y0=1
yexact0= 1
y.append(y0)
for i in range (1,len(t)):
  y[0]=1
  y.append(a)
  y[i]=(1-delta_t*k)*y[i-1]

yexact.append(yexact0)
for i in range (1,len(t)):
  t[0]=0
  yexact.append(a)

  yexact[i]=math.exp(-t[i])

plt.figure(figsize=(12,8))
figure=plt.subplot()
figure.set_title('dy/dt = -y, y(0) = 1,FORWARD EULER METHOD')
figure.set_xlabel('t-axis')
figure.set_ylabel('y-axis')
plt.plot(t,y,'bo--',label='Approximate')
plt.plot(t,yexact,'go--',label='yexact')
plt.legend(loc='lower right')
plt.show()
```

```python
#Exercise 28
#Backward Euler Method
import numpy as npy
import matplotlib.pyplot as plt
from math import *
f=lambda t:2*(math.exp(-2*t))
inip=0
finp=2
delta_t=0.1
k=2
N_t=int((finp-inip)/delta_t)
t=npy.arange(0,2+delta_t,delta_t)
y=[]
yexact=[]
t0=0
y0=2
yexact0=2
a=1
y.append(y0)
for i in range (1,len(t)):
  y[0]=2
  y.append(a)
  y[i]=(y[i-1]/(1+delta_t*k))

yexact.append(yexact0)
for i in range (1,len(t)):
  yexact[0]=2
  yexact.append(a)
  yexact[i]=2*(math.exp(-2*t[i]))

plt.figure(figsize=(12,8))
figure=plt.subplot()
figure.set_title('dy/dt = -2y, y(0) = 2,BACKWARD EULER METHOD')
figure.set_xlabel('t-axis')
figure.set_ylabel('y-axis')
plt.plot(t,y,'bo--',label='Approximate')
plt.plot(t,yexact,'go--',label='yexact')
plt.legend(loc='lower right')
plt.show()
```
